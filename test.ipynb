```python
import json

with open('tabelog_#tokyo_restaurants_score_above_3.5.json', 'r') as f:
    tabelog_info = json.load(f)

# Initial info list
all_restaurant = []
value_l = []

for key, value in tabelog_info.items():
    if 'chinese' not in key:
        value = [dict(item, type=key) for item in value]
        value = [dict(item, chinese_type=tabelog_info[f'{key}_chinese']) for item in value]

        all_restaurant+=value

        value_l += value



value_l
```
